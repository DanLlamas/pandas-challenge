```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
Total_schools = len(school_data["school_name"].unique())
total_students=len(student_data["student_name"])
Total_budget=school_data["budget"].sum()
Average_Math_Score=student_data["math_score"].mean()
Average_Reading_Score=student_data["reading_score"].mean()

overall_passing_math= school_data_complete.query('math_score >=70')["School ID"].count()/ total_students*100
overall_passing_read= school_data_complete.query('reading_score >=70')["School ID"].count()/ total_students*100
overall_passing_rate= (overall_passing_math + overall_passing_read)/2


Total_schools, total_students, Total_budget, Average_Math_Score,Average_Reading_Score, overall_passing_math, overall_passing_read, overall_passing_rate
```

```python
DF_District_Summary=pd.DataFrame({"Total Schools":[Total_schools],"Total Students":[total_students],
                                  "Total Budget":[Total_budget],"Average Math Score":[Average_Math_Score],
                                  "Average Reading Score":[Average_Reading_Score],
                                  "% Passing Math":[overall_passing_math],
                                  "% Passing Reading":[overall_passing_read],
                                  "% Overall Passing Rate":[overall_passing_rate]})
DF_District_Summary ["Total Students"]= DF_District_Summary ["Total Students"].map("{:,.0f}".format)
DF_District_Summary ["Total Budget"]= DF_District_Summary ["Total Budget"].round(2).map("${:,.2f}".format)

DF_District_Summary.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_data = school_data_complete[["School ID", "school_name", "type", "size", "budget", "Student ID", 
                                                 "student_name","gender", "grade", "reading_score","math_score"]].copy()

school_data_group=school_data.groupby(["school_name","type"])

school_data_group.head()
```

```python
total_students_sch = school_data_group["student_name"].count()
total_budget_sch = school_data_group["budget"].mean()
budget_pp_sch =  (total_budget_sch / total_students_sch)
avg_math_sch = school_data_group["math_score"].mean()
avg_reading_sch = school_data_group["reading_score"].mean()
passing_math_sch = school_data_complete.query('math_score >=70')["School ID"].count()/ total_students_sch
#school_data_complete["math_score"]/school_data_complete["Student ID"]
passing_reading_sch = school_data_complete.query('reading_score >=70')["School ID"].count()/ total_students_sch 
#school_data_complete["reading_score"]/school_data_complete["Student ID"]
overall_pass_rate_sch = ((passing_math_sch + passing_reading_sch)/2)

df_School_Summary=pd.DataFrame({"Total Students":total_students_sch,
                                "Total School Budget":total_budget_sch,
                                "Per Student Budget":budget_pp_sch,
                               "Average Math Score":avg_math_sch,
                                "Average Reading Score":avg_reading_sch,
                                "%  Passing Math":passing_math_sch,
                                "% Passing Reading":passing_reading_sch,
                                "% Overall Passing Rate":overall_pass_rate_sch})



df_School_Summary=df_School_Summary.round(2)

df_School_Summary["Total Students"]=df_School_Summary["Total Students"].map("{:,.2f}".format)
df_School_Summary["Total School Budget"]=df_School_Summary["Total School Budget"].map("${:,.2f}".format)
df_School_Summary["Per Student Budget"]=df_School_Summary["Per Student Budget"].map("${:,.2f}".format)

DF_Schools=df_School_Summary.sort_values(by=["% Overall Passing Rate"],ascending=False)

DF_Schools.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
DF_Schools=df_School_Summary.sort_values(by=["% Overall Passing Rate"],ascending=True)

DF_Schools.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
df_School_Summary_List = df_School_Summary.copy()

DF_Budget=df_School_Summary_List[["Per Student Budget","Average Math Score","Average Reading Score","%  Passing Math","% Passing Reading","% Overall Passing Rate"]]

DF_Budget
```

```python
DF_Budget_Max=df_School_Summary_List["Per Student Budget"].max()
DF_Budget_Min=df_School_Summary_List["Per Student Budget"].min()

DF_Budget_Max,DF_Budget_Min
```

```python
# Sample bins. Feel free to create your own bins.
bins = [0, 585, 615, 645, 675]
names = ["$578-584", "$585-615", "$615-645", "$645-675"]
```

```python
DF_Budget["Spending ranges (Per Student)"] = pd.cut(budget_pp_sch, bins, labels=names)

math_scores = DF_Budget.groupby(["Spending ranges (Per Student)"]).mean()["Average Math Score"]
reading_scores = DF_Budget.groupby(["Spending ranges (Per Student)"]).mean()["Average Reading Score"]
passing_math_sc = DF_Budget.groupby(["Spending ranges (Per Student)"]).mean()["%  Passing Math"]
passing_reading_sc = DF_Budget.groupby(["Spending ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_sc = (passing_math_sc + passing_reading_sc)/2

DF_score = pd.DataFrame({"Average Math Score":math_scores,"Average Reading Score":reading_scores,
                        "% Passing Math":passing_math_sc,"% Passing Reading":passing_reading_sc,
                        "% Overall Passing Rate": overall_passing_sc})

DF_score
```
